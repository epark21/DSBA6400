```
import pandas as pd
x = pd.DataFrame(x)
udates=x['date'].unique()
x['date'] = pandas.to_datetime(x['date'])

```

```
filled_df = (x.set_index('date')
             .groupby('tic')
             .apply(lambda d: d.reindex(pd.date_range(min(x.date),
                                                      max(x.date),
                                                      freq='D')))
             .drop('tic', axis=1)
             .reset_index('tic')
             .fillna(0))


```
